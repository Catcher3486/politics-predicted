```python
import sklearn
import pandas as pd
import numpy as np  

#Linear regression with one variable for testing
from sklearn import linear_model
from sklearn.linear_model import LinearRegression

df = pd.read_csv("US_Election_dataset_v1.csv", sep = ",", header=0)
cols = ["county","state","Density per square km","2020 Democrat vote %"]
df_subset= df[cols]



y = np.array(df_subset["2020 Democrat vote %"])
x = np.array(df_subset["Density per square km"]).reshape((-1, 1))


model=LinearRegression()
model.fit(x,y)

#R² test to see how much variance we're accounting for
r_sq = model.score(x, y)
print("total variance accounted for:")
print({r_sq})
```
