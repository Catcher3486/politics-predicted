```python
#importing libraries and reading dataframe
import pandas as pd
import numpy as np
import sklearn as sk
df = pd.read_csv("US_Election_dataset_v1.csv", sep = ",", header=0)

cols = ["county","state","Density per square km","2020 Democrat vote %"]
df_subset= df[cols]
print(df_subset)
```

```python
##basic descriptive statistics to expand on

#percentile distribution of variables in my dataset
#df_subset.describe()

#extracting republican & democratic votes by state
results_by_state = (
    df.groupby("state", as_index=False)[
        ["2020 Republican vote %", "2020 Democrat vote %"]
    ]
    .mean()
)

#adding a column for overall voter turnout
results_by_state["total_vote_%"] = (
    results_by_state["2020 Republican vote %"] 
    + results_by_state["2020 Democrat vote %"]
)

results_by_state.sort_values(by="total_vote_%",ascending=False)


```
