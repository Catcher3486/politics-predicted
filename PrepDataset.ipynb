```python
#importing libraries and reading dataframe
import pandas as pd
import numpy as np
import sklearn as sk
df = pd.read_csv("US_Election_dataset_v1.csv", sep = ",", header=0)

cols = ["county","state","Density per square km","2020 Democrat vote %"]
df_subset= df[cols]
print(df_subset)
```

```python
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
y = np.array(df_subset["2020 Democrat vote %"])
x = np.array(df_subset["Density per square km"]).reshape((-1, 1))
print(y.shape)

model=LinearRegression()
model.fit(x,y)

r_sq = model.score(x, y)
print("coefficient of determination: {r_sq}")
```
