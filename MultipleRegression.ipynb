```python
import sklearn
import pandas as pd
import numpy as np  

#Linear regression with one variable for testing
from sklearn import linear_model
from sklearn.linear_model import LinearRegression

df = pd.read_csv("US_Election_dataset_v1.csv", sep = ",", header=0)
cols = ["county","state","Gini Index","Area in square Km", "2020 Democrat vote %"]
df_subset = df[cols]


#unlike the linear regression, let's try to create the model without numpy arrays, and keep it in pandas dataframes
y = df_subset["2020 Democrat vote %"]
x = df_subset[["Gini Index","Area in square Km"]]


model=LinearRegression()
model.fit(x,y)

#R² test to see how much variance we're accounting for
r_sq = model.score(x, y)*100
print("total variance accounted for:")
print(f"{r_sq:.2f}%")
```
